```python
import os
import re
import time
import random
from groq import Groq
from dataclasses import dataclass, field
from typing import List, Tuple
from math import log, sqrt
```

```python
from dotenv import load_dotenv
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)

# Initialize LLaMA3-8B model
model_name = "llama3-8b-8192"  # Replace with actual model name if different

def LLM_response(prompt, system_prompt="", temperature=0, max_tokens=16384, model_name=model_name):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": f"{system_prompt}",
            },
            {
                "role": "user",
                "content": f"{prompt}",
            }
        ],
        model=model_name,
        temperature=temperature,
        max_tokens=max_tokens,
    )

    return chat_completion.choices[0].message.content
```

```python
def direct_response(question: str) -> str:
        prompt = (
            "Aswer the following question.\n"
            f"Question: {question}\n\n"
            "Let's think step by step"
            )
        direct_response = LLM_response(prompt)
        return direct_response
```

```python
question = "Kevin Kangaroo begins hopping on a number line at 0. He wants to get to 1, but he can hop only 1/3 of the distance. Each hop tires him out so that he continues to hop 1/3 of the remaining distance. How far has he hopped after five hops? Express your answer as a common fraction."
correct_answer = "211/243"
response = direct_response(question)
print(response)
```

```python
def critique(question: str, draft_answer: str) -> str:
        prompt = (
            f"Question: {question}\n"
            f"Draft answer: {draft_answer}\n"
            "Please provide a comprehensive critique of the draft answer by addressing the following points:\n\n"
            "1. Accuracy and Correctness:\n"
            "   - Is the answer factually correct? Explain why or why not.\n"
            "   - Are there any logical inconsistencies or errors in reasoning?\n"
            "   - Does the answer fully address all aspects of the question?\n\n"
            "2. Verification Methods:\n"
            "   - Suggest at least three different ways to verify the correctness of the answer.\n"
            "   - Apply these verification methods and report your findings.\n\n"
            "3. Clarity and Coherence:\n"
            "   - Is the answer well-structured and easy to understand?\n"
            "   - Are there any ambiguous or vague statements that need clarification?\n\n"
            "4. Completeness:\n"
            "   - Does the answer cover all necessary information?\n"
            "   - Are there any important points or perspectives missing?\n\n"
            "5. Flaws and Weaknesses:\n"
            "   - Identify and explain every flaw in the draft answer, no matter how minor.\n"
            "   - Hold the answer to the highest possible standard in your critique.\n\n"
            "6. Improvement Recommendations:\n"
            "   - Provide specific, actionable recommendations to improve each aspect of the answer.\n"
            "   - Suggest additional information or context that could enhance the answer.\n\n"
            "7. Step-by-Step Analysis:\n"
            "   - Break down your critique into clear, logical steps.\n"
            "   - Ensure each step of your analysis is thorough and well-explained.\n\n"
            "Remember: Do not provide a revised answer. Focus solely on critiquing and providing recommendations for improvement.\n"
            "Let's work through this carefully to ensure we have the right critique."
            )
        critique = LLM_response(prompt)
        return critique
```

```python
# draft_answer = "I don't know"
critique_reponse = critique(question, response)
print(critique_reponse)
```

```python
def improve_answer(question: str, draft_answer: str, critique: str) -> str:
        prompt = (
            f"Task: Improve the draft answer based on the given critique while ensuring relevance to the original question.\n\n"
            f"Original Question: {question}\n"
            f"Draft Answer: {draft_answer}\n"
            f"Critique: {critique}\n\n"
            "Please provide an improved answer following this format:\n\n"
            "1. Reasoning Process:\n"
            "   - Analyze the critique and identify areas for improvement\n"
            "   - Provide a step-by-step explanation of how you'll address each point\n"
            "   - Ensure your reasoning directly relates to the original question\n\n"
            "2. Verification:\n"
            "   - Fact-check the information in the improved answer\n"
            "   - Cite reliable sources if applicable\n"
            "   - Ensure all statements are accurate and well-supported\n\n"
            "3. Final Answer:\n"
            "   - Provide a concise, clear, and improved answer\n"
            "   - Incorporate the insights from your reasoning and verification\n"
            "   - Ensure the answer directly addresses the original question\n\n"
            "Remember to be concise, clear, and directly address all points raised in the critique.\n"
            "Your improved answer should be more accurate, comprehensive, and relevant to the original question."
            )
        improved_answer = LLM_response(prompt)
        return improved_answer.strip()
```

```python
improve_answer = improve_answer(question, response, critique_reponse)
print(improve_answer)
```

```python
def rate_answer(question: str, answer: str) -> float:
        prompt = (
            f"Question: {question}\n"
            f"Answer: {answer}\n"
            "Please analyze the answer based on the following criteria:\n"
            "1. Accuracy of information\n"
            "2. Completeness of the response\n"
            "3. Relevance to the question\n"
            "4. Clarity and coherence\n"
            "5. Use of evidence or examples (if applicable)\n\n"
            "Provide a detailed critique addressing each of these aspects. Be specific about strengths and weaknesses, "
            "and point out any factual errors or misconceptions. Do not suggest improvements or provide a corrected answer.\n\n"
            "After your critique, rate the answer on a scale of 0 to 100, where:\n"
            "0-20: Poor (major flaws, largely incorrect or irrelevant)\n"
            "21-40: Below Average (significant issues, partially relevant or correct)\n"
            "41-60: Average (some flaws, mostly relevant and partially correct)\n"
            "61-80: Good (minor issues, relevant and mostly correct)\n"
            "81-100: Excellent (minimal to no issues, highly relevant and correct)\n\n"
            "Your response should follow this format:\n"
            "Critique: <detail critique>\n"
            "Rating: <rating>"
            )
        rating_response = LLM_response(prompt)
        # Extract the rating 
        try:
            match = re.search(r"Rating:\s*(\d+)", rating_response)
            if match:
                rating = int(match.group(1))
                if rating > 95:
                    rating = 95
                rating = float(rating)/100
            else:
                raise ValueError("Rating not found in the response") 
        except Exception as e:
            print(f"Error extracting rating: {e}")
            print(f"Rating response was {rating_response}")
            rating = 0.0
        return rating
```

```python
def generate_expert_methods(question: str, other_expert_idea: str = "") -> str:
    # Construct the initial part of the prompt
    prompt = (
        f"You are an expert tasked with providing step-by-step instructions to analyze and outline methods for solving this question:\n"
        f"Question: {question}\n\n"
        "Instructions:\n"
        "1. Analyze the type of problem presented in the question.\n"
        "2. Brainstorm your thinking process and outline the steps you would take to answer the question. Create a relationships map for better visualize is encouraged\n"
        "3. Provide a clear step-by-step guide on how you would approach the problem. Do not solve the problem completely or provide any computations.\n"
        "4. Your response should be in the following format:\n"
        "   Expert:\n"
        "   Instruction: <Step-by-step instructions>\n"
    )
    
    # Add other expert ideas if they exist
    if other_expert_idea:
        prompt += (
            "\nAdditional Context:\n"
            "You are working in a group of experts. Another expert has proposed the following approach:\n"
            f"Other Expert's Idea: {other_expert_idea}\n\n"
            "Your task is to provide a different approach from the one mentioned above, ensuring a diverse range of solutions.\n"
        )
        
    # Simulate a response from an LLM (Language Model)
    expert_methods = LLM_response(prompt, temperature=1.0)
    return expert_methods

# Main execution
# question = "There are 3 men in the room. One person is cooking. A person is playing chess. What are the other people doing?"
question = "If I hang 5 shirts outside and it takes them 5 hours to dry, how long would it take to dry 30 shirts?"
other_expert_idea = ""

# Collect expert methods in a list
expert_methods_list = []
for _ in range(3):
    expert_methods = generate_expert_methods(question, other_expert_idea)
    expert_methods_list.append(expert_methods)
    # Update the string with all previous expert ideas
    other_expert_idea += expert_methods + " "

# Print the collected expert methods
for i,method in enumerate(expert_methods_list):
    print(f"====================== Expert {i+1} ======================\n")
    print(method)
    print("=========================================================\n")
```

```python
def generate_answer(question: str, expert_idea: str) -> str:
    # Construct the initial part of the prompt
    prompt = (
        f"Question: {question}\n\n"
        "Please provide a comprehensive answer with detailed reasoning and analysis. The answer is definitely in the question. Follow this structure from the expert:\n"
        f"Expert idea: {expert_idea}\n\n"
        "Let's work this out in a step by step way to be sure we have the right answer.\n"
        "Your response should be in the following format:\n"
        "Reasoning step: <Step-by-step reasoning>\n"
        "Final answer: <final answer>\n"
    )
    
    # Simulate a response from an LLM (Language Model)
    answer = LLM_response(prompt)
    return answer

for i,method in enumerate(expert_methods_list):
    answer = generate_answer(question, method)
    print(f"====================== Answer {i+1} ======================\n")
    print(answer)
    print("=======================================================\n")
```
